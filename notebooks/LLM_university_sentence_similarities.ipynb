```python
texts = ["I like to be in my house",
         "I enjoy staying home",
         "the isotope 238u decays to 206pb"]
```

# Embedding

```python
#get the embedding from the response
embeddings = response.embeddings.float
```

```python
#unpack the three sentences in a list using the embedding
[sentence1, sentence2, sentence3] = embeddings

for sentence in [sentence1, sentence2, sentence3]:
  print(f"sentence-embedding: {sentence}\n")
```

# Dot Product Similarity

```python
print(f"dot product similarity between sentence1 and sentence2: {np.dot(sentence1, sentence2)}")
print(f"dot product similarity between sentence1 and sentence3: {np.dot(sentence1, sentence3)}")
print(f"dot product similarity between sentence2 and sentence3: {np.dot(sentence2, sentence3)}")
```

# Cosine Similarity

```python
print(f"Cosine Similarity between sentence1 and sentence2: {cosine_similarity([sentence1], [sentence2])[0][0]}")
print(f"Cosine Similarity between sentence1 and sentence3: {cosine_similarity([sentence1], [sentence3])[0][0]}")
print(f"Cosine Similarity between sentence2 and sentence3: {cosine_similarity([sentence2], [sentence3])[0][0]}")
```
